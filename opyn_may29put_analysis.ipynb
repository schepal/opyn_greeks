```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("opyn_data_compiled.csv", index_col='time', parse_dates=True)
```

```python
contract = "ETH-MAY29-200P"
(df.iv*100).plot()
plt.title("Implied Volatility: " + contract)
plt.ylabel("Implied Volatility (%)")
plt.xlabel("Date");
```

```python
contract = "ETH-MAY29-200P"
(-df.delta).plot()
plt.title("Delta: " + contract)
plt.ylabel("Put Delta (Negative)")
plt.xlabel("Date");
```

```python
contract = "ETH-MAY29-200P"
(df.gamma).plot()
plt.title("Gamma: " + contract)
plt.ylabel("Gamma")
plt.xlabel("Date");
```

```python
contract = "ETH-MAY29-200P"
(df.theta).plot()
plt.title("Theta: " + contract)
plt.ylabel("Theta")
plt.xlabel("Date");
```

```python
plt.scatter(df.iv*100, df.gamma, alpha=0.05)
plt.title("Gamma vs. Implied Volatility " + contract)
plt.ylabel("Gamma")
plt.xlabel("Implied Volatility (%)");
```
